# Inflammation analysis
## Plot mean, max, min in a single figure

```python
import numpy
import matplotlib.pyplot as plt

#Import data
data = numpy.loadtxt(fname="data/inflammation-01.csv", delimiter =',')

#Set up figure
fig = plt.figure(figsize=(10.0,3.0))

# add subplots
axes1 = fig.add_subplot(1, 3, 1) #nrows, ncols, pos
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

axes1.plot(numpy.mean(data, axis=0))
axes2.plot(numpy.max(data,axis=0))
axes3.plot(numpy.min(data,axis=0))

axes1.set_ylabel('mean')
axes2.set_ylabel('max')
axes3.set_ylabel('min')

fig.savefig('inflammation.png')
```

## Loop analysis over filenames

```python
import glob
import numpy
import matplotlib.pyplot as plt

filenames = sorted(glob.glob('data/inflammation*.csv'))

for filename in filenames[:3]:
    #Import data
    data = numpy.loadtxt(fname=filename, delimiter =',')

    #Set up figure
    fig = plt.figure(figsize=(10.0,3.0))

    # add subplots
    axes1 = fig.add_subplot(1, 3, 1) #nrows, ncols, pos
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.plot(numpy.mean(data, axis=0))
    axes2.plot(numpy.max(data,axis=0))
    axes3.plot(numpy.min(data,axis=0))

    axes1.set_ylabel('mean')
    axes2.set_ylabel('max')
    axes3.set_ylabel('min')
```

# Refactoring code in functions

```python
def visualize(filename):
    #Import data
    data = numpy.loadtxt(fname=filename, delimiter =',')

    #Set up figure
    fig = plt.figure(figsize=(10.0,3.0))

    # add subplots
    axes1 = fig.add_subplot(1, 3, 1) #nrows, ncols, pos
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.plot(numpy.mean(data, axis=0))
    axes2.plot(numpy.max(data,axis=0))
    axes3.plot(numpy.min(data,axis=0))

    axes1.set_ylabel('mean')
    axes2.set_ylabel('max')
    axes3.set_ylabel('min')
```

```python
import numpy
import matplotlib.pyplot as plt

visualize('data/inflammation-01.csv')
```
